```python
age = int(input("請輸入年齡"))
if age>=18:
    print("滿18了")
    print("可以考駕照了")
else:
    print("未滿18歲")
    print(f"再等{18-age}年，才可以考駕照")
print("程式結束")
```

```python
#巢狀判斷

age = int(input("請輸入年齡"))
if age<18:
    print("您還未成年")
else: #判斷條件相當於18<=age
    if age<20: #判斷條件相當於18<=age<20
        print("您已成年，但還不能投票")
    else: #判斷條件相當於age>=20，所以age>=18也>=20
        print("您是個成年人且可投票")
print("程式結束")
```

```python
#多項判斷

age = int(input("請輸入年齡"))
print(f"您的年齡為{age}")
if age<18:
    print("您還未成年")
elif age<=21:
    print("您已成年，未滿22歲，尚不行投票")
elif age<22: #判斷條件相當於18<=age<20
    print("您已成年，但還不能投票")
#多項判斷可以建立多個程式區塊，但是每次都只會執行一個
#而且只會執行第一個出現TRUE判斷式的區塊，然後就會直接跳出
#此規則適用於目前所有的程式語言

elif age<21:
    print("您已成年，未滿22歲，尚不行投票")
#上述這段程式碼若放在age<22之後，即使填入19歲也無法執行
#因為21歲已經先被判定為<22歲，所以先執行<22歲的程式區塊:print("您已成年，但還不能投票")

else: #判斷條件相當於age>=18也>=22
    print("您是個成年人且可投票")
print("程式結束")
```

```python
#if只有一個判斷式，所以必須要從旁邊開始判斷起
score = 55
if score>=90:
    print("優")
elif score>=80: #目前條件已經為<90的前提下再做判斷
    print("甲")
elif score>=70: #目前條件已經為<80的前提下再做判斷
    print("乙")
elif score>=60: #目前條件已經為<70的前提下再做判斷
    print("丙")
else:           #目前條件已經為<60的前提下再做判斷
    print("丁")
```

```python
#上段程式也可以反著寫
score = 90
if score<60:
    print("丁")
elif score<70: #目前條件已經為>=60的前提下再做判斷
    print("丙")
elif score<80: #目前條件已經為>=70的前提下再做判斷
    print("乙")
elif score<90: #目前條件已經為>=80的前提下再做判斷
    print("甲")
else:           #目前條件已經為>=90的前提下再做判斷
    print("優")
```

```python
#邏輯運算子，讓兩個以上的bool變成一個bool
#and是二元邏輯
#嚴格得到TRUE
presure = 78
print(presure)
if presure >= 90 and presure < 120:
    print("正常")
else:
    print("不正常")
```

```python
#or是容易得到TRUE
presure = 100
print(presure)
if presure < 90 or presure >= 120:
    print("不正常")
else:
    print("正常")
```

```python
#not是反向
#not (TRUE)-->FALSE ; not (FALSE)-->TRUE
presure = 150
print(presure)
if not(presure < 90 or presure >= 120): 
    print("正常")
else:
    print("不正常")
```

```python
#elif後面也可直接以數學式表示(可讀性較高)，但大概只有python才可以這樣寫
score = 95
print(score)
if score>=90:
    print("優")
elif 90>score>=80: #目前條件已經為<90的前提下再做判斷
    print("甲")
elif 80>score>=70: #目前條件已經為<80的前提下再做判斷
    print("乙")
elif 70>score>=60: #目前條件已經為<70的前提下再做判斷
    print("丙")
else:           #目前條件已經為<60的前提下再做判斷
    print("丁")
```
