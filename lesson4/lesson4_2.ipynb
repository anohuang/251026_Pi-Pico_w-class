```python
age = int(input("請輸入年齡"))
if age>=18:
    print("滿18了")
    print("可以考駕照了")
else:
    print("未滿18歲")
    print(f"再等{18-age}年，才可以考駕照")
print("程式結束")
```

```python
#巢狀判斷

age = int(input("請輸入年齡"))
if age<18:
    print("您還未成年")
else: #判斷條件相當於18<=age
    if age<20: #判斷條件相當於18<=age<20
        print("您已成年，但還不能投票")
    else: #判斷條件相當於age>=20，所以age>=18也>=20
        print("您是個成年人且可投票")
print("程式結束")
```
