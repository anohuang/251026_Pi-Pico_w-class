```python
#如果不知道迴圈的次數，就用while
#while用來判斷bool: TRUE時會反覆執行程式，FALSE時會自動跳出
import random

while random.randint(1,10) != 5: 
    print(random.randint(1,10))
    print("不等於5")
print("等於5")
```

```python
import random

pp=random.randint(1,10) #這樣寫的話，亂數生成只會執行一次
while pp != 5: #迴圈裡不能用pp代替，因為數值不會變化，所以會跳不出來
               #因為已經進入了無限迴圈，所以要加寫break
    print(pp)
    print("不等於5")
    break
print("等於5")
```

```python
import random

while True: #此為無限迴圈，會永遠跳不出去
    pp = random.randint(1,10)
    if pp == 5:
        break #使用無限迴圈時一定要搭配break
              #碰到break，就會跳出while，執行後面ROOT區的程式
    print(pp)
    print("不等於5") #單向選擇，else可寫可不寫
print("等於5")
print(pp)
```
