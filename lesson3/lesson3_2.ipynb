```python
name = input("請輸入你的名字")
height = float(input("請輸入你的身高"))
weight = float(input("請輸入你的體重"))
BMI = weight / ((height/100)**2)
print ("%s的身高是%.2f公分" %(name, height))
print ("%s的體重是%.1f公斤" %(name, weight))
print ("{:s}的BMI是{:d}" .format(name, int(BMI)))
```
